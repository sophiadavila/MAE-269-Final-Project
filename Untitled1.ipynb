```python
"""
Three-Link Swimmer — Path Tracking with Gait Switching
=======================================================

Architecture:
  Desired Path
       ↓
  Path Tracking Controller  ←── Cycle-averaged Position / Heading
       ↓                                   ↑
  Desired Swimming Direction      3-Link Swimmer (one full cycle)
       ↓                                   ↑
  Gait Selector               Disturbance (known cross-flow)
       ↓
  Joint Kinematics (held fixed for one gait cycle)

Key design decisions
--------------------
* Gait is held CONSTANT for one full oscillation cycle (T = 1 s).
  At the end of each cycle the controller reads the CYCLE-AVERAGED
  position and heading, eliminating the rapid θ oscillation that
  polluted the previous continuous controller.
* Path: straight line y = 0 (fish swims in the −x direction).
  path_closest projects onto the full y=0 line without clamping.
* Five discrete gaits parameterised by body-bias angle δ:
    α₁(t) = δ + A·sin(ωt),   α₂(t) = δ + A·sin(ωt + π/2)
  δ = 0  → symmetric, straight forward
  δ > 0  → body curves +y  → fish curves toward +y (left turn)
  δ < 0  → body curves −y  → fish curves toward −y (right turn)
* Proportional controller:  δ_demand = −K_ct · e_ct − K_θ · e_θ
  where e_ct = y  and  e_θ = θ_avg (desired heading = 0).
"""

from sympy import *
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt

# ─────────────────────────────────────────────────────────────────────────────
# 1.  SYMBOLIC VARIABLES
# ─────────────────────────────────────────────────────────────────────────────
t = symbols("t", real=True)
l, m, I_link, c = symbols("l m I c", positive=True, real=True)
U_flow   = symbols("U_flow",   real=True)
U_flow_y = symbols("U_flow_y", real=True)

x     = Function("x")(t)
y     = Function("y")(t)
theta = Function("theta")(t)
a1    = Function("a1")(t)
a2    = Function("a2")(t)

q    = [x, y, theta, a1, a2]
q_dt = [diff(v, t) for v in q]
n    = 5

# ─────────────────────────────────────────────────────────────────────────────
# 2.  SE(3) HELPERS
# ─────────────────────────────────────────────────────────────────────────────
def R_z(ang):
    return Matrix([[cos(ang), -sin(ang), 0],
                   [sin(ang),  cos(ang), 0],
                   [0,         0,        1]])

def make_T(R, p):
    return Matrix([
        [R[0,0], R[0,1], R[0,2], p[0,0]],
        [R[1,0], R[1,1], R[1,2], p[1,0]],
        [R[2,0], R[2,1], R[2,2], p[2,0]],
        [0,      0,      0,      1      ]
    ])

# ─────────────────────────────────────────────────────────────────────────────
# 3.  LINK TRANSFORMS   [Front] --α₁-- [Base] --α₂-- [Tail]
# ─────────────────────────────────────────────────────────────────────────────
T_base  = make_T(R_z(theta), Matrix([[x], [y], [0]]))

T_front = (T_base
           @ make_T(Matrix.eye(3), Matrix([[-l/2], [0], [0]]))
           @ make_T(R_z(a1),       Matrix([[0],    [0], [0]]))
           @ make_T(Matrix.eye(3), Matrix([[-l/2], [0], [0]])))

T_tail  = (T_base
           @ make_T(Matrix.eye(3), Matrix([[l/2], [0], [0]]))
           @ make_T(R_z(a2),       Matrix([[0],   [0], [0]]))
           @ make_T(Matrix.eye(3), Matrix([[l/2], [0], [0]])))

# ─────────────────────────────────────────────────────────────────────────────
# 4.  LINK POSITIONS AND TRANSLATIONAL JACOBIANS
# ─────────────────────────────────────────────────────────────────────────────
r_base  = Matrix([T_base[0, 3],  T_base[1, 3]])
r_front = simplify(Matrix([T_front[0, 3], T_front[1, 3]]))
r_tail  = simplify(Matrix([T_tail[0,  3], T_tail[1,  3]]))

J_base  = r_base.jacobian(q)
J_front = simplify(r_front.jacobian(q))
J_tail  = simplify(r_tail.jacobian(q))

print("── Jacobians ──────────────────────")
print("J_front:"); display(J_front)
print("J_tail:");  display(J_tail)

# ─────────────────────────────────────────────────────────────────────────────
# 5.  KINETIC ENERGY AND MASS MATRIX
# ─────────────────────────────────────────────────────────────────────────────
qd = Matrix(q_dt)

def link_KE(J_v, omega_sym):
    v = J_v @ qd
    return Rational(1,2)*m*(v.T @ v)[0,0] + Rational(1,2)*I_link*omega_sym**2

T_kin = expand(
    link_KE(J_base,  q_dt[2])            +
    link_KE(J_front, q_dt[2] + q_dt[3]) +
    link_KE(J_tail,  q_dt[2] + q_dt[4])
)

M_mat = simplify(Matrix([
    [diff(diff(T_kin, q_dt[i]), q_dt[j]) for j in range(n)]
    for i in range(n)
]))

print("\n── Mass matrix M(q) ────────────────"); display(M_mat)

# ─────────────────────────────────────────────────────────────────────────────
# 6.  CORIOLIS / CENTRIFUGAL VECTOR
# ─────────────────────────────────────────────────────────────────────────────
q_ddt_syms = symbols("xdd ydd thdd a1dd a2dd")
h_vec = Matrix.zeros(n, 1)

for i in range(n):
    dT_i  = diff(T_kin, q_dt[i])
    ddt_i = sum(
        diff(dT_i, q[j])    * q_dt[j] +
        diff(dT_i, q_dt[j]) * q_ddt_syms[j]
        for j in range(n)
    )
    h_vec[i] = simplify(
        ddt_i.subs(dict(zip(q_ddt_syms, [0]*n))) - diff(T_kin, q[i])
    )

print("\n── Coriolis vector h(q,q_dot) ──────"); display(h_vec)

# ─────────────────────────────────────────────────────────────────────────────
# 7.  FLOW FIELD AND GENERALISED HYDRODYNAMIC FORCES
# ─────────────────────────────────────────────────────────────────────────────
def flow_field(px, py):
    return Matrix([U_flow, U_flow_y])

def generalized_drag_force(J_v, r_i):
    vf_i  = flow_field(r_i[0], r_i[1])
    v_i   = J_v @ qd
    vrel  = v_i - vf_i
    F_i   = -c * vrel
    return J_v.T @ F_i

Q_flow = simplify(
    generalized_drag_force(J_base,  r_base)  +
    generalized_drag_force(J_front, r_front) +
    generalized_drag_force(J_tail,  r_tail)
)

print("\n── Q_flow ──────────────────────────"); display(Q_flow)

# ─────────────────────────────────────────────────────────────────────────────
# 8.  LAMBDIFY
# ─────────────────────────────────────────────────────────────────────────────
params_sym = [l, m, I_link, c, U_flow, U_flow_y]
M_fn  = lambdify([q, params_sym],        M_mat,  "numpy")
h_fn  = lambdify([q, q_dt, params_sym],  h_vec,  "numpy")
Qf_fn = lambdify([q, q_dt, params_sym],  Q_flow, "numpy")

print("\nSymbolic derivation complete. Starting numerical simulations...\n")

# ═════════════════════════════════════════════════════════════════════════════
#  PATH TRACKING WITH GAIT SWITCHING
# ═════════════════════════════════════════════════════════════════════════════

# ─────────────────────────────────────────────────────────────────────────────
# 9.  PHYSICAL PARAMETERS
# ─────────────────────────────────────────────────────────────────────────────
l_val = 1.0
m_val = 0.1
I_val = m_val * l_val**2 / 12.0
c_val = 1.0

def make_params(Ux=0.0, Uy=0.0):
    return [l_val, m_val, I_val, c_val, Ux, Uy]

# ─────────────────────────────────────────────────────────────────────────────
# 10. GAIT LIBRARY
#
#   α₁(t) = δ + A·sin(ωt)
#   α₂(t) = δ + A·sin(ωt + π/2)
#
#   δ controls the body-bias (average curvature) and therefore the
#   net turning tendency over one full cycle.
# ─────────────────────────────────────────────────────────────────────────────
W_GAIT  = 2.0 * np.pi       # [rad/s]
PHI_G   = np.pi / 2         # tail–front phase offset
A_GAIT  = np.pi / 4         # oscillation amplitude [rad]
T_GAIT  = 2.0*np.pi / W_GAIT  # period = 1.0 s

GAITS = [
    ('sharp_right',  -0.50),
    ('gentle_right', -0.25),
    ('forward',       0.00),
    ('gentle_left',  +0.25),
    ('sharp_left',   +0.50),
]
GAIT_NAMES  = [g[0] for g in GAITS]
GAIT_DELTAS = [g[1] for g in GAITS]

def gait_kinematics(tv, delta):
    """(α₁, α₂, α̇₁, α̇₂, α̈₁, α̈₂) for bias δ at time tv."""
    a1v  =  delta + A_GAIT * np.sin(W_GAIT * tv)
    a2v  =  delta + A_GAIT * np.sin(W_GAIT * tv + PHI_G)
    a1d  =          A_GAIT * W_GAIT    * np.cos(W_GAIT * tv)
    a2d  =          A_GAIT * W_GAIT    * np.cos(W_GAIT * tv + PHI_G)
    a1dd =         -A_GAIT * W_GAIT**2 * np.sin(W_GAIT * tv)
    a2dd =         -A_GAIT * W_GAIT**2 * np.sin(W_GAIT * tv + PHI_G)
    return a1v, a2v, a1d, a2d, a1dd, a2dd

# ─────────────────────────────────────────────────────────────────────────────
# 11. LOW-LEVEL DYNAMICS (one call, gait fixed from outside)
# ─────────────────────────────────────────────────────────────────────────────
def dynamics(tv, state, delta, Ux, Uy):
    xv, yv, thv, xd, yd, thd = state
    p_vals = make_params(Ux=Ux, Uy=Uy)
    a1v, a2v, a1d, a2d, a1dd, a2dd = gait_kinematics(tv, delta)
    qc  = [xv, yv, thv, a1v, a2v]
    qdc = [xd, yd, thd, a1d, a2d]
    M_  = np.array(M_fn(qc, p_vals),       dtype=float)
    h_  = np.array(h_fn(qc, qdc, p_vals),  dtype=float).flatten()
    Qf_ = np.array(Qf_fn(qc, qdc, p_vals), dtype=float).flatten()
    rhs        = Qf_[0:3] - h_[0:3] - M_[0:3, 3:5] @ np.array([a1dd, a2dd])
    q_ddot_pas = np.linalg.solve(M_[0:3, 0:3], rhs)
    return [xd, yd, thd, *q_ddot_pas]

# ─────────────────────────────────────────────────────────────────────────────
# 12. DESIRED PATH  (straight line y = 0)
#
#   The fish naturally swims in the −x direction with this gait setup,
#   so the path extends along the x-axis in both directions.
#   For a straight-line path the closest point is simply [x_fish, 0].
#   No clamping needed — the fish should go wherever it goes in x.
# ─────────────────────────────────────────────────────────────────────────────
def path_closest(pos):
    """
    Closest point on y = 0 to pos = [x, y].
    Returns (closest_point, desired_heading).
    Desired heading = 0 because when the fish swims straight along y=0,
    its base-link angle θ ≈ 0 (base is horizontal, front pointing −x).
    """
    r_star  = np.array([pos[0], 0.0])   # project onto y=0, any x allowed
    psi_d   = 0.0                        # desired base-link orientation
    return r_star, psi_d

# ─────────────────────────────────────────────────────────────────────────────
# 13. PATH TRACKING CONTROLLER
#
#   Inputs:  cycle-averaged position [x̄, ȳ] and heading θ̄
#   Outputs: continuous turning demand δ_demand
#
#   Control law:
#     e_ct = ȳ               (cross-track error: + = above path)
#     e_θ  = θ̄ − ψ_d        (heading error:     + = CCW from desired)
#     δ    = −K_ct·e_ct − K_θ·e_θ
#
#   When ȳ > 0 → δ < 0 → right-curving gait → fish curves back to y=0 ✓
#   When θ̄ > 0 → δ < 0 → right-curving gait → fish straightens ✓
# ─────────────────────────────────────────────────────────────────────────────
K_CT      = 1.0     # cross-track gain  [rad / m]
K_TH      = 0.8     # heading gain      [rad / rad]
DELTA_MAX = 0.50    # clamp = sharpest gait

def path_controller(pos_avg, theta_avg):
    r_star, psi_d = path_closest(pos_avg)
    e_ct = pos_avg[1] - r_star[1]                    # = ȳ
    e_th = theta_avg - psi_d                          # = θ̄
    e_th = (e_th + np.pi) % (2.0 * np.pi) - np.pi   # wrap to (−π, π]
    delta_demand = -K_CT * e_ct - K_TH * e_th
    delta_demand = np.clip(delta_demand, -DELTA_MAX, DELTA_MAX)
    return delta_demand, e_ct, e_th

# ─────────────────────────────────────────────────────────────────────────────
# 14. GAIT SELECTOR  (nearest-neighbour quantisation)
# ─────────────────────────────────────────────────────────────────────────────
def gait_selector(delta_demand):
    idx = int(np.argmin([abs(d - delta_demand) for _, d in GAITS]))
    return GAITS[idx]   # (name, delta_value)

# ─────────────────────────────────────────────────────────────────────────────
# 15. DISTURBANCE MODEL
#   List of (t_start, t_end, Ux, Uy) intervals.
# ─────────────────────────────────────────────────────────────────────────────
def get_disturbance(tv, disturbances):
    for t0, t1, Ux, Uy in disturbances:
        if t0 <= tv < t1:
            return Ux, Uy
    return 0.0, 0.0

# ─────────────────────────────────────────────────────────────────────────────
# 16. CYCLE-BASED SIMULATION WITH PATH TRACKING
#
#   The gait is held fixed for exactly one oscillation cycle (T_GAIT = 1 s).
#   At the end of each cycle:
#     1. Compute the CIRCULAR MEAN of θ over the cycle  → removes oscillation
#     2. Use mean position + mean heading to update the path controller
#     3. Select the nearest discrete gait for the next cycle
#
#   This mirrors a real gait-switching fish: it completes one full stroke
#   then decides whether to continue straight or turn.
# ─────────────────────────────────────────────────────────────────────────────
def simulate_path_tracking(t_end, s0, disturbances, pts_per_cycle=200):
    """
    Returns:
      t_arr      : (N,)   time stamps
      y_arr      : (6, N) states  [x, y, θ, ẋ, ẏ, θ̇]
      delta_arr  : (N,)   gait bias active at each time step
      gait_arr   : (N,)   gait name at each time step
      ctrl_log   : list of dicts with per-cycle controller data
    """
    state    = np.array(s0, dtype=float)
    t_now    = 0.0
    delta    = 0.0          # start with forward gait
    gname    = 'forward'

    t_list     = []
    y_list     = []
    delta_list = []
    gait_list  = []
    ctrl_log   = []         # record per-cycle controller info

    cycle_idx = 0
    while t_now < t_end - 1e-10:
        t_next = min(t_now + T_GAIT, t_end)
        t_seg  = np.linspace(t_now, t_next, pts_per_cycle)
        d_hold = delta  # hold this gait for the full cycle

        # --- Integrate one gait cycle ----------------------------------------
        def ode(tv, s):
            Ux, Uy = get_disturbance(tv, disturbances)
            return dynamics(tv, s, d_hold, Ux, Uy)

        sol = solve_ivp(ode, [t_now, t_next], state,
                        t_eval=t_seg, rtol=1e-8, atol=1e-8, method='RK45')

        # Store (exclude last point to avoid duplicate at cycle boundary)
        t_list.append(sol.t[:-1])
        y_list.append(sol.y[:, :-1])
        delta_list.extend([d_hold] * (len(sol.t) - 1))
        gait_list.extend([gname]   * (len(sol.t) - 1))

        state = sol.y[:, -1]
        t_now = t_next

        # --- Controller update: use cycle-averaged state ----------------------
        x_avg   = np.mean(sol.y[0])
        y_avg   = np.mean(sol.y[1])
        # Circular mean of heading (robust to wrap-around near 0/±π)
        th_avg  = np.angle(np.mean(np.exp(1j * sol.y[2])))

        delta_demand, e_ct, e_th = path_controller(
            np.array([x_avg, y_avg]), th_avg
        )
        gname, delta = gait_selector(delta_demand)

        ctrl_log.append({
            'cycle':  cycle_idx,
            't':      t_now,
            'x_avg':  x_avg,
            'y_avg':  y_avg,
            'th_avg': th_avg,
            'e_ct':   e_ct,
            'e_th':   e_th,
            'delta_d': delta_demand,
            'delta':  delta,
            'gname':  gname,
        })
        cycle_idx += 1

    # Append final point
    t_list.append(np.array([t_now]))
    y_list.append(state.reshape(6, 1))
    delta_list.append(delta)
    gait_list.append(gname)

    t_arr     = np.concatenate(t_list)
    y_arr     = np.concatenate(y_list, axis=1)
    delta_arr = np.array(delta_list)

    return t_arr, y_arr, delta_arr, gait_list, ctrl_log

# ─────────────────────────────────────────────────────────────────────────────
# 17. DIAGNOSTIC — verify natural swimming direction (no controller, δ = 0)
# ─────────────────────────────────────────────────────────────────────────────
print("Diagnostic: natural swimming direction with forward gait (δ=0)…")
t_diag = np.linspace(0, 3 * T_GAIT, 600)
sol_diag = solve_ivp(
    lambda tv, s: dynamics(tv, s, 0.0, 0.0, 0.0),
    [0, 3 * T_GAIT], [0.0]*6, t_eval=t_diag, rtol=1e-8, atol=1e-8
)
dx_nat = sol_diag.y[0, -1]
dy_nat = sol_diag.y[1, -1]
print(f"  After 3 cycles:  Δx = {dx_nat:.4f} m,  Δy = {dy_nat:.4f} m")
print(f"  Natural swimming angle ≈ {np.degrees(np.arctan2(dy_nat, dx_nat)):.1f}°")

# ─────────────────────────────────────────────────────────────────────────────
# 18. SIMULATIONS
#
#   A) No disturbance — baseline path tracking
#   B) Single cross-flow pulse  Uy = 0.8 m/s during t ∈ [3, 6] s
#   C) Two opposing pulses      (push up, then push down)
# ─────────────────────────────────────────────────────────────────────────────
T_END = 20.0    # 20 gait cycles total
s0    = [0.0] * 6

# Scenario A
print("\nScenario A: no disturbance…")
tA, yA, dA, gnA, logA = simulate_path_tracking(T_END, s0, [])
print(f"  x_final={yA[0,-1]:.3f}  y_final={yA[1,-1]:.4f}")

# Scenario B — single cross-flow
dist_B = [(4.0, 7.0, 0.0, 0.8)]
print("Scenario B: Uy=0.8 m/s during t∈[4,7] s…")
tB, yB, dB, gnB, logB = simulate_path_tracking(T_END, s0, dist_B)
print(f"  x_final={yB[0,-1]:.3f}  y_final={yB[1,-1]:.4f}")

# Scenario C — two opposing pulses
dist_C = [(3.0, 6.0, 0.0, 0.8), (11.0, 14.0, 0.0, -0.8)]
print("Scenario C: two opposing cross-flow pulses…")
tC, yC, dC, gnC, logC = simulate_path_tracking(T_END, s0, dist_C)
print(f"  x_final={yC[0,-1]:.3f}  y_final={yC[1,-1]:.4f}")

# ─────────────────────────────────────────────────────────────────────────────
# 19. EXTRACT PER-CYCLE CONTROLLER DATA FOR PLOTTING
# ─────────────────────────────────────────────────────────────────────────────
def extract_log(log):
    t_c     = np.array([d['t']       for d in log])
    e_ct_c  = np.array([d['e_ct']    for d in log])
    e_th_c  = np.array([d['e_th']    for d in log])
    delta_c = np.array([d['delta']   for d in log])
    delta_d = np.array([d['delta_d'] for d in log])
    gname_c = [d['gname'] for d in log]
    return t_c, e_ct_c, e_th_c, delta_c, delta_d, gname_c

tA_c, eA_ct, eA_th, dA_c, dA_d, gnA_c = extract_log(logA)
tB_c, eB_ct, eB_th, dB_c, dB_d, gnB_c = extract_log(logB)
tC_c, eC_ct, eC_th, dC_c, dC_d, gnC_c = extract_log(logC)

# ─────────────────────────────────────────────────────────────────────────────
# 20. PLOTS
# ─────────────────────────────────────────────────────────────────────────────
fig, axes = plt.subplots(3, 2, figsize=(14, 13))
fig.suptitle(
    "Three-Link Swimmer — Path Tracking with Gait Switching\n"
    "(cycle-based controller, Δt_ctrl = 1 gait cycle)",
    fontsize=13, fontweight='bold'
)

# ── Plot 1: XY trajectories ──────────────────────────────────────────────────
ax = axes[0, 0]
x_lo = min(yA[0].min(), yB[0].min(), yC[0].min()) * 1.05
x_hi = max(yA[0].max(), yB[0].max(), yC[0].max()) * 1.05
ax.axhline(0, color='k', ls='--', lw=2.5, zorder=5, label='Desired path (y=0)')
ax.plot(yA[0], yA[1], 'b-',  lw=1.5, label='A: no disturbance')
ax.plot(yB[0], yB[1], 'r-',  lw=1.5, label='B: Uy pulse t∈[4,7] s')
ax.plot(yC[0], yC[1], 'g-',  lw=1.5, label='C: two pulses')
ax.plot(0, 0, 'ko', ms=6, zorder=10, label='Start')
ax.set_xlabel('x [m]'); ax.set_ylabel('y [m]')
ax.set_title('Swimmer Trajectories')
ax.legend(fontsize=8); ax.grid(True); ax.axis('equal')

# ── Plot 2: Cross-track error per cycle (B) ──────────────────────────────────
ax = axes[0, 1]
ax.step(tB_c, eB_ct, 'r-', where='post', lw=2.0, label='e_ct (Scenario B)')
ax.axhline(0, color='k', ls='--', lw=1.2)
for (t0, t1, _, _) in dist_B:
    ax.axvspan(t0, t1, alpha=0.20, color='orange', label='Disturbance on')
ax.set_xlabel('t [s]  (sampled once per cycle)'); ax.set_ylabel('e_ct [m]')
ax.set_title('Cross-Track Error — Scenario B\n(cycle-averaged position)')
ax.legend(fontsize=8); ax.grid(True)

# ── Plot 3: Heading error per cycle (B) ──────────────────────────────────────
ax = axes[1, 0]
ax.step(tB_c, np.degrees(eB_th), 'm-', where='post', lw=2.0,
        label='e_θ (Scenario B)')
ax.axhline(0, color='k', ls='--', lw=1.2)
for (t0, t1, _, _) in dist_B:
    ax.axvspan(t0, t1, alpha=0.20, color='orange', label='Disturbance on')
ax.set_xlabel('t [s]  (sampled once per cycle)'); ax.set_ylabel('e_θ [°]')
ax.set_title('Heading Error — Scenario B\n(circular mean over cycle)')
ax.legend(fontsize=8); ax.grid(True)

# ── Plot 4: Gait switching (B) ───────────────────────────────────────────────
ax = axes[1, 1]
ax.step(tB_c, dB_c, 'b-', where='post', lw=2.5, label='Selected gait δ')
ax.plot(tB_c, dB_d, 'r--', lw=1.0, alpha=0.7, label='Demanded δ (continuous)')
ax.axhline(0, color='k', ls=':', lw=0.6)
for (t0, t1, _, _) in dist_B:
    ax.axvspan(t0, t1, alpha=0.20, color='orange', label='Disturbance on')
for name, val in GAITS:
    ax.axhline(val, color='gray', ls=':', lw=0.4)
ax.set_yticks(GAIT_DELTAS)
ax.set_yticklabels(GAIT_NAMES, fontsize=8)
ax.set_xlabel('t [s]'); ax.set_ylabel('Active gait')
ax.set_title('Gait Switching — Scenario B\n(updates once per cycle)')
ax.legend(fontsize=7); ax.grid(True)

# ── Plot 5: y-position vs time (all scenarios, full resolution) ──────────────
ax = axes[2, 0]
ax.axhline(0, color='k', ls='--', lw=2.5, label='Desired y = 0')
ax.plot(tA, yA[1], 'b-', lw=1.2, label='A: no disturbance')
ax.plot(tB, yB[1], 'r-', lw=1.2, label='B: single pulse')
ax.plot(tC, yC[1], 'g-', lw=1.2, label='C: two pulses')
for (t0, t1, _, _) in dist_C:
    ax.axvspan(t0, t1, alpha=0.10, color='orange')
ax.set_xlabel('t [s]'); ax.set_ylabel('y [m]')
ax.set_title('Lateral Position vs Time\n(path tracking & recovery)')
ax.legend(fontsize=8); ax.grid(True)

# ── Plot 6: Cross-track error all three scenarios (per-cycle) ────────────────
ax = axes[2, 1]
ax.axhline(0, color='k', ls='--', lw=1.5)
ax.step(tA_c, eA_ct, 'b-', where='post', lw=1.8, label='A: no disturbance')
ax.step(tB_c, eB_ct, 'r-', where='post', lw=1.8, label='B: single pulse')
ax.step(tC_c, eC_ct, 'g-', where='post', lw=1.8, label='C: two pulses')
for (t0, t1, _, _) in dist_C:
    ax.axvspan(t0, t1, alpha=0.10, color='orange')
ax.set_xlabel('t [s]'); ax.set_ylabel('e_ct [m]')
ax.set_title('Cross-Track Error — All Scenarios\n(cycle-averaged)')
ax.legend(fontsize=8); ax.grid(True)

plt.tight_layout()
plt.savefig("path_tracking_gait_switching.png", dpi=120, bbox_inches="tight")
plt.show()
print("\nDone. Plot saved to path_tracking_gait_switching.png")
```
