```python
"""
Three-Link Swimmer — Path Tracking with Gait Switching
=======================================================

Architecture (ChatGPT-suggested pipeline):

  Desired Path
       ↓
  Path Tracking Controller  ←──  Position / Orientation
       ↓                                  ↑
  Desired Swimming Direction      3-Link Swimmer
       ↓                                  ↑
  Gait Selector               Disturbance (known flow)
       ↓
  Joint Kinematics (prescribed gait)

Gait parameterisation — bias-angle method:
  α₁(t) = δ + A·sin(ωt)
  α₂(t) = δ + A·sin(ωt + π/2)
  δ = 0  → straight / forward gait
  δ > 0  → curves toward +y (left turn)
  δ < 0  → curves toward −y (right turn)

Five discrete gaits:
  sharp_right (δ=−0.5), gentle_right (δ=−0.25), forward (δ=0),
  gentle_left (δ=+0.25), sharp_left (δ=+0.5)

Desired path: straight line along x-axis (y=0, heading ψ=0).
Disturbances: known cross-flow intervals (t_start, t_end, Ux, Uy).

Generalised coordinates:  q = [x, y, θ, α₁, α₂]
"""

from sympy import *
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt

# ─────────────────────────────────────────────────────────────────────────────
# 1.  SYMBOLIC VARIABLES
# ─────────────────────────────────────────────────────────────────────────────
t = symbols("t", real=True)

l, m, I_link, c = symbols("l m I c", positive=True, real=True)
U_flow   = symbols("U_flow",   real=True)
U_flow_y = symbols("U_flow_y", real=True)

x     = Function("x")(t)
y     = Function("y")(t)
theta = Function("theta")(t)
a1    = Function("a1")(t)
a2    = Function("a2")(t)

q    = [x, y, theta, a1, a2]
q_dt = [diff(v, t) for v in q]
n    = 5

# ─────────────────────────────────────────────────────────────────────────────
# 2.  SE(3) HELPERS
# ─────────────────────────────────────────────────────────────────────────────
def R_z(ang):
    return Matrix([[cos(ang), -sin(ang), 0],
                   [sin(ang),  cos(ang), 0],
                   [0,         0,        1]])

def make_T(R, p):
    return Matrix([
        [R[0,0], R[0,1], R[0,2], p[0,0]],
        [R[1,0], R[1,1], R[1,2], p[1,0]],
        [R[2,0], R[2,1], R[2,2], p[2,0]],
        [0,      0,      0,      1      ]
    ])

# ─────────────────────────────────────────────────────────────────────────────
# 3.  LINK TRANSFORMS
# ─────────────────────────────────────────────────────────────────────────────
T_base  = make_T(R_z(theta), Matrix([[x], [y], [0]]))

T_front = (T_base
           @ make_T(Matrix.eye(3), Matrix([[-l/2], [0], [0]]))
           @ make_T(R_z(a1),       Matrix([[0],    [0], [0]]))
           @ make_T(Matrix.eye(3), Matrix([[-l/2], [0], [0]])))

T_tail  = (T_base
           @ make_T(Matrix.eye(3), Matrix([[l/2], [0], [0]]))
           @ make_T(R_z(a2),       Matrix([[0],   [0], [0]]))
           @ make_T(Matrix.eye(3), Matrix([[l/2], [0], [0]])))

# ─────────────────────────────────────────────────────────────────────────────
# 4.  LINK POSITIONS AND TRANSLATIONAL JACOBIANS
# ─────────────────────────────────────────────────────────────────────────────
r_base  = Matrix([T_base[0, 3],  T_base[1, 3]])
r_front = simplify(Matrix([T_front[0, 3], T_front[1, 3]]))
r_tail  = simplify(Matrix([T_tail[0,  3], T_tail[1,  3]]))

J_base  = r_base.jacobian(q)
J_front = simplify(r_front.jacobian(q))
J_tail  = simplify(r_tail.jacobian(q))

print("── Jacobians ──────────────────────")
print("J_front:"); display(J_front)
print("J_tail:");  display(J_tail)

# ─────────────────────────────────────────────────────────────────────────────
# 5.  KINETIC ENERGY AND MASS MATRIX
# ─────────────────────────────────────────────────────────────────────────────
qd = Matrix(q_dt)

def link_KE(J_v, omega_sym):
    v = J_v @ qd
    return Rational(1,2)*m*(v.T @ v)[0,0] + Rational(1,2)*I_link*omega_sym**2

T_kin = expand(
    link_KE(J_base,  q_dt[2])            +
    link_KE(J_front, q_dt[2] + q_dt[3]) +
    link_KE(J_tail,  q_dt[2] + q_dt[4])
)

M_mat = simplify(Matrix([
    [diff(diff(T_kin, q_dt[i]), q_dt[j]) for j in range(n)]
    for i in range(n)
]))

print("\n── Mass matrix M(q) ────────────────"); display(M_mat)

# ─────────────────────────────────────────────────────────────────────────────
# 6.  CORIOLIS / CENTRIFUGAL VECTOR
# ─────────────────────────────────────────────────────────────────────────────
q_ddt_syms = symbols("xdd ydd thdd a1dd a2dd")
h_vec = Matrix.zeros(n, 1)

for i in range(n):
    dT_i  = diff(T_kin, q_dt[i])
    ddt_i = sum(
        diff(dT_i, q[j])    * q_dt[j] +
        diff(dT_i, q_dt[j]) * q_ddt_syms[j]
        for j in range(n)
    )
    h_vec[i] = simplify(
        ddt_i.subs(dict(zip(q_ddt_syms, [0]*n))) - diff(T_kin, q[i])
    )

print("\n── Coriolis vector h(q,q_dot) ──────"); display(h_vec)

# ─────────────────────────────────────────────────────────────────────────────
# 7.  FLOW FIELD AND GENERALISED HYDRODYNAMIC FORCES
# ─────────────────────────────────────────────────────────────────────────────
def flow_field(px, py):
    return Matrix([U_flow, U_flow_y])

def generalized_drag_force(J_v, r_i):
    vf_i  = flow_field(r_i[0], r_i[1])
    v_i   = J_v @ qd
    vrel  = v_i - vf_i
    F_i   = -c * vrel
    return J_v.T @ F_i

Q_flow = simplify(
    generalized_drag_force(J_base,  r_base)  +
    generalized_drag_force(J_front, r_front) +
    generalized_drag_force(J_tail,  r_tail)
)

print("\n── Q_flow ──────────────────────────"); display(Q_flow)

# ─────────────────────────────────────────────────────────────────────────────
# 8.  LAMBDIFY
# ─────────────────────────────────────────────────────────────────────────────
params_sym = [l, m, I_link, c, U_flow, U_flow_y]

M_fn  = lambdify([q, params_sym],        M_mat,  "numpy")
h_fn  = lambdify([q, q_dt, params_sym],  h_vec,  "numpy")
Qf_fn = lambdify([q, q_dt, params_sym],  Q_flow, "numpy")

print("\nSymbolic derivation complete. Starting numerical simulations...\n")

# ═════════════════════════════════════════════════════════════════════════════
#  PATH TRACKING WITH GAIT SWITCHING  (sections 9–18)
# ═════════════════════════════════════════════════════════════════════════════

# ─────────────────────────────────────────────────────────────────────────────
# 9.  PHYSICAL PARAMETERS
# ─────────────────────────────────────────────────────────────────────────────
l_val = 1.0
m_val = 0.1
I_val = m_val * l_val**2 / 12.0
c_val = 1.0

def make_params(Ux=0.0, Uy=0.0):
    return [l_val, m_val, I_val, c_val, Ux, Uy]

# ─────────────────────────────────────────────────────────────────────────────
# 10. GAIT LIBRARY
#
#   Each gait is fully defined by a single bias angle δ.
#   Joint kinematics:
#     α₁(t) = δ + A·sin(ωt)
#     α₂(t) = δ + A·sin(ωt + π/2)
#
#   δ = 0   → symmetric oscillation → straight forward swimming
#   δ > 0   → body curves toward +y → left turn
#   δ < 0   → body curves toward −y → right turn
#
#   The five discrete gaits span the useful turning range:
# ─────────────────────────────────────────────────────────────────────────────
W_GAIT = 2.0 * np.pi   # gait frequency [rad/s]
PHI_G  = np.pi / 2     # tail–front phase offset [rad]
A_GAIT = np.pi / 4     # oscillation amplitude [rad]

# Ordered from most-right to most-left
GAITS = [
    ('sharp_right',  -0.50),
    ('gentle_right', -0.25),
    ('forward',       0.00),
    ('gentle_left',  +0.25),
    ('sharp_left',   +0.50),
]
GAIT_NAMES  = [g[0] for g in GAITS]
GAIT_DELTAS = [g[1] for g in GAITS]

def gait_kinematics(tv, delta):
    """
    Return (α₁, α₂, α̇₁, α̇₂, α̈₁, α̈₂) for bias δ at time tv.
    Note: derivatives of the bias term are zero (δ is constant per gait).
    """
    a1v  =  delta + A_GAIT * np.sin(W_GAIT * tv)
    a2v  =  delta + A_GAIT * np.sin(W_GAIT * tv + PHI_G)
    a1d  =          A_GAIT * W_GAIT    * np.cos(W_GAIT * tv)
    a2d  =          A_GAIT * W_GAIT    * np.cos(W_GAIT * tv + PHI_G)
    a1dd =         -A_GAIT * W_GAIT**2 * np.sin(W_GAIT * tv)
    a2dd =         -A_GAIT * W_GAIT**2 * np.sin(W_GAIT * tv + PHI_G)
    return a1v, a2v, a1d, a2d, a1dd, a2dd

# ─────────────────────────────────────────────────────────────────────────────
# 11. DESIRED PATH
#
#   Straight line along the x-axis: y_d = 0, ψ_d = 0.
#   To change path shape, override path_point() and path_heading() below.
#   Example curved paths are provided as comments.
# ─────────────────────────────────────────────────────────────────────────────
def path_point(s):
    """World position on the desired path at arc-length s."""
    # Straight line:
    return np.array([s, 0.0])
    # Sinusoidal:  return np.array([s, 0.5 * np.sin(0.5 * s)])
    # Circle r=3:  return np.array([3*np.cos(s/3), 3*np.sin(s/3)])

def path_heading(s):
    """Desired heading angle ψ_d [rad] at arc-length s."""
    # Straight line:
    return 0.0
    # Sinusoidal:  return np.arctan2(0.5*0.5*np.cos(0.5*s), 1.0)

def path_closest(pos):
    """
    For a straight-line path along x:
      s* = x  (projection of pos onto x-axis, clamped ≥ 0)
    For other paths, replace with a numerical minimizer.
    """
    s_star = max(0.0, pos[0])
    return s_star, path_point(s_star), path_heading(s_star)

# ─────────────────────────────────────────────────────────────────────────────
# 12. PATH TRACKING CONTROLLER
#
#   Proportional controller on two error signals:
#     e_ct : cross-track error  (signed lateral deviation from path) [m]
#     e_th : heading error      (current heading minus desired)      [rad]
#
#   Control output:  δ = −K_CT·e_ct − K_TH·e_th,  clipped to [−δ_max, δ_max]
#
#   This δ is the turning demand passed to the gait selector.
# ─────────────────────────────────────────────────────────────────────────────
K_CT      = 1.5    # cross-track gain  [rad / m]
K_TH      = 1.0    # heading gain      [rad / rad]
DELTA_MAX = 0.50   # maximum bias angle = sharpest available gait [rad]

def path_controller(pos, theta):
    """
    Args:
      pos   : [x, y] current position
      theta : current heading [rad]
    Returns:
      delta_demand : continuous turning demand [rad], clipped
      e_ct         : cross-track error [m]
      e_th         : heading error [rad]
    """
    s_star, r_star, psi_d = path_closest(pos)

    # Cross-track error: positive = swimmer is above (left of) the path
    e_ct = pos[1] - r_star[1]

    # Heading error, wrapped to (−π, π]
    e_th = theta - psi_d
    e_th = (e_th + np.pi) % (2.0 * np.pi) - np.pi

    delta_demand = -K_CT * e_ct - K_TH * e_th
    delta_demand = np.clip(delta_demand, -DELTA_MAX, DELTA_MAX)

    return delta_demand, e_ct, e_th

# ─────────────────────────────────────────────────────────────────────────────
# 13. GAIT SELECTOR
#
#   Maps the continuous turning demand δ_demand to the nearest
#   discrete gait in the library (nearest-neighbour quantisation).
#   This is the step that makes the controller a true "gait switcher"
#   rather than a continuous actuator.
# ─────────────────────────────────────────────────────────────────────────────
def gait_selector(delta_demand):
    """
    Returns (gait_name, delta_value) of the discrete gait
    closest to delta_demand.
    """
    idx = int(np.argmin([abs(d - delta_demand) for _, d in GAITS]))
    return GAITS[idx]

# ─────────────────────────────────────────────────────────────────────────────
# 14. DISTURBANCE MODEL
#
#   Known disturbances: list of tuples (t_start, t_end, Ux, Uy).
#   During [t_start, t_end) the background flow is (Ux, Uy) m/s,
#   which acts as a cross-current pushing the swimmer off the path.
# ─────────────────────────────────────────────────────────────────────────────
def get_disturbance(tv, disturbances):
    """Return (Ux, Uy) at time tv given the disturbance schedule."""
    for t0, t1, Ux, Uy in disturbances:
        if t0 <= tv < t1:
            return Ux, Uy
    return 0.0, 0.0

# ─────────────────────────────────────────────────────────────────────────────
# 15. ODE WITH PATH TRACKING + GAIT SWITCHING
#
#   State: s = [x, y, θ, ẋ, ẏ, θ̇]
#   At each evaluation:
#     1. Compute disturbance → p_vals (flow parameters)
#     2. Path tracking controller → δ_demand
#     3. Gait selector            → discrete δ
#     4. Joint kinematics         → (α₁, α₂, α̇₁, α̇₂, α̈₁, α̈₂)
#     5. Passive-DOF EOM         → (ẍ, ÿ, θ̈)
# ─────────────────────────────────────────────────────────────────────────────
def dynamics_tracking(tv, state, disturbances):
    xv, yv, thv, xd, yd, thd = state

    # Step 1: disturbance
    Ux_d, Uy_d = get_disturbance(tv, disturbances)
    p_vals = make_params(Ux=Ux_d, Uy=Uy_d)

    # Step 2: path tracking controller
    delta_demand, _, _ = path_controller(np.array([xv, yv]), thv)

    # Step 3: gait selection (discrete switching)
    _, delta = gait_selector(delta_demand)

    # Step 4: joint kinematics for selected gait
    a1v, a2v, a1d, a2d, a1dd, a2dd = gait_kinematics(tv, delta)

    # Step 5: solve passive EOM
    qc  = [xv, yv, thv, a1v, a2v]
    qdc = [xd, yd, thd, a1d, a2d]

    M_  = np.array(M_fn(qc, p_vals),       dtype=float)
    h_  = np.array(h_fn(qc, qdc, p_vals),  dtype=float).flatten()
    Qf_ = np.array(Qf_fn(qc, qdc, p_vals), dtype=float).flatten()

    rhs        = Qf_[0:3] - h_[0:3] - M_[0:3, 3:5] @ np.array([a1dd, a2dd])
    q_ddot_pas = np.linalg.solve(M_[0:3, 0:3], rhs)

    return [xd, yd, thd, *q_ddot_pas]

# ─────────────────────────────────────────────────────────────────────────────
# 16. SIMULATIONS
#
#   Three scenarios, all starting at rest at the origin:
#     A) No disturbance  — baseline path tracking
#     B) Single cross-flow pulse (Uy = 0.8 m/s, t ∈ [2, 3.5] s)
#     C) Two opposing pulses  (push up then push down)
# ─────────────────────────────────────────────────────────────────────────────
T_END  = 8.0
N_EVAL = 8000
t_span = [0.0, T_END]
t_eval = np.linspace(0.0, T_END, N_EVAL)
s0     = [0.0] * 6   # at rest at origin, heading θ = 0

# ── Scenario A: no disturbance ───────────────────────────────────────────────
print("Scenario A: no disturbance…")
sol_a = solve_ivp(
    lambda tv, s: dynamics_tracking(tv, s, []),
    t_span, s0, t_eval=t_eval, rtol=1e-7, atol=1e-7, method='RK45'
)
print(f"  x_final = {sol_a.y[0,-1]:.3f} m,  y_final = {sol_a.y[1,-1]:.4f} m")

# ── Scenario B: single cross-flow ────────────────────────────────────────────
dist_b = [(2.0, 3.5, 0.0, 0.8)]
print("Scenario B: Uy = 0.8 m/s during t ∈ [2, 3.5] s…")
sol_b = solve_ivp(
    lambda tv, s: dynamics_tracking(tv, s, dist_b),
    t_span, s0, t_eval=t_eval, rtol=1e-7, atol=1e-7, method='RK45'
)
print(f"  x_final = {sol_b.y[0,-1]:.3f} m,  y_final = {sol_b.y[1,-1]:.4f} m")

# ── Scenario C: two opposing disturbances ────────────────────────────────────
dist_c = [
    (1.5, 3.0,  0.0,  0.8),   # push upward (+y)
    (5.0, 6.5,  0.0, -0.8),   # push downward (−y)
]
print("Scenario C: two opposing cross-flow pulses…")
sol_c = solve_ivp(
    lambda tv, s: dynamics_tracking(tv, s, dist_c),
    t_span, s0, t_eval=t_eval, rtol=1e-7, atol=1e-7, method='RK45'
)
print(f"  x_final = {sol_c.y[0,-1]:.3f} m,  y_final = {sol_c.y[1,-1]:.4f} m")

# ─────────────────────────────────────────────────────────────────────────────
# 17. POST-PROCESS — errors and gait selections along Scenario B
# ─────────────────────────────────────────────────────────────────────────────
e_ct_b     = np.zeros(N_EVAL)
e_th_b     = np.zeros(N_EVAL)
delta_d_b  = np.zeros(N_EVAL)
gait_sel_b = np.zeros(N_EVAL)   # stores the selected δ value

for i in range(N_EVAL):
    xv, yv, thv = sol_b.y[0, i], sol_b.y[1, i], sol_b.y[2, i]
    dd, e_ct, e_th = path_controller(np.array([xv, yv]), thv)
    _, gdelta = gait_selector(dd)
    e_ct_b[i]     = e_ct
    e_th_b[i]     = e_th
    delta_d_b[i]  = dd
    gait_sel_b[i] = gdelta

# ─────────────────────────────────────────────────────────────────────────────
# 18. PLOTS
# ─────────────────────────────────────────────────────────────────────────────
x_max  = max(sol_a.y[0].max(), sol_b.y[0].max(), sol_c.y[0].max()) * 1.05
x_path = np.array([0.0, x_max])

fig, axes = plt.subplots(3, 2, figsize=(14, 13))
fig.suptitle("Three-Link Swimmer — Path Tracking with Gait Switching",
             fontsize=14, fontweight='bold')

# ── Plot 1: XY trajectories ──────────────────────────────────────────────────
ax = axes[0, 0]
ax.plot(x_path, [0, 0], 'k--', lw=2.5, zorder=5, label='Desired path (y=0)')
ax.plot(sol_a.y[0], sol_a.y[1], 'b-',  lw=1.5, label='A: no disturbance')
ax.plot(sol_b.y[0], sol_b.y[1], 'r-',  lw=1.5, label='B: Uy pulse t∈[2,3.5]')
ax.plot(sol_c.y[0], sol_c.y[1], 'g-',  lw=1.5, label='C: two disturbances')
ax.set_xlabel('x [m]'); ax.set_ylabel('y [m]')
ax.set_title('Swimmer Trajectories')
ax.legend(fontsize=8); ax.grid(True); ax.axis('equal')

# ── Plot 2: Cross-track error — Scenario B ───────────────────────────────────
ax = axes[0, 1]
ax.plot(t_eval, e_ct_b, 'r-', lw=1.5, label='Cross-track error')
ax.axhline(0, color='k', ls='--', lw=1.2)
for (t0, t1, _, _) in dist_b:
    ax.axvspan(t0, t1, alpha=0.20, color='orange', label='Disturbance on')
ax.set_xlabel('t [s]'); ax.set_ylabel('e_ct [m]')
ax.set_title('Cross-Track Error — Scenario B')
ax.legend(fontsize=8); ax.grid(True)

# ── Plot 3: Heading error — Scenario B ───────────────────────────────────────
ax = axes[1, 0]
ax.plot(t_eval, np.degrees(e_th_b), 'm-', lw=1.5, label='Heading error')
ax.axhline(0, color='k', ls='--', lw=1.2)
for (t0, t1, _, _) in dist_b:
    ax.axvspan(t0, t1, alpha=0.20, color='orange', label='Disturbance on')
ax.set_xlabel('t [s]'); ax.set_ylabel('e_θ [°]')
ax.set_title('Heading Error — Scenario B')
ax.legend(fontsize=8); ax.grid(True)

# ── Plot 4: Gait switching — Scenario B ──────────────────────────────────────
ax = axes[1, 1]
ax.step(t_eval, gait_sel_b, 'b-', lw=2.0, where='post', label='Selected gait δ', zorder=3)
ax.plot(t_eval, delta_d_b,  'r-', lw=0.9, alpha=0.6, label='Demanded δ (continuous)')
ax.axhline(0, color='k', ls='--', lw=0.6)
for (t0, t1, _, _) in dist_b:
    ax.axvspan(t0, t1, alpha=0.20, color='orange', label='Disturbance on')
# Gait level annotations
for name, val in GAITS:
    ax.axhline(val, color='gray', ls=':', lw=0.5)
ax.set_yticks(GAIT_DELTAS)
ax.set_yticklabels(GAIT_NAMES, fontsize=7)
ax.set_xlabel('t [s]'); ax.set_ylabel('Selected gait')
ax.set_title('Gait Switching Over Time — Scenario B')
ax.legend(fontsize=7); ax.grid(True)

# ── Plot 5: Lateral position vs time (all three scenarios) ───────────────────
ax = axes[2, 0]
ax.axhline(0, color='k', ls='--', lw=2.5, label='Desired  y = 0')
ax.plot(t_eval, sol_a.y[1], 'b-', lw=1.5, label='A: no disturbance')
ax.plot(t_eval, sol_b.y[1], 'r-', lw=1.5, label='B: single pulse')
ax.plot(t_eval, sol_c.y[1], 'g-', lw=1.5, label='C: two pulses')
# Shade all disturbance windows
all_dists = dist_b + dist_c
for (t0, t1, _, _) in all_dists:
    ax.axvspan(t0, t1, alpha=0.10, color='orange')
ax.set_xlabel('t [s]'); ax.set_ylabel('y [m]')
ax.set_title('Lateral Position vs Time (path tracking recovery)')
ax.legend(fontsize=8); ax.grid(True)

# ── Plot 6: Heading vs time — Scenario B and C ───────────────────────────────
ax = axes[2, 1]
ax.axhline(0, color='k', ls='--', lw=2.5, label='Desired  θ = 0')
ax.plot(t_eval, np.degrees(sol_b.y[2]), 'r-', lw=1.5, label='B: heading θ')
ax.plot(t_eval, np.degrees(sol_c.y[2]), 'g-', lw=1.5, label='C: heading θ')
for (t0, t1, _, _) in dist_b + dist_c:
    ax.axvspan(t0, t1, alpha=0.12, color='orange')
ax.set_xlabel('t [s]'); ax.set_ylabel('θ [°]')
ax.set_title('Heading Angle vs Time (convergence to 0°)')
ax.legend(fontsize=8); ax.grid(True)

plt.tight_layout()
plt.savefig("path_tracking_gait_switching.png", dpi=120, bbox_inches="tight")
plt.show()
print("\nDone. Plot saved to path_tracking_gait_switching.png")
```
